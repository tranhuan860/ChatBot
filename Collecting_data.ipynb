```python

from g4f.client import Client

G4F_PROXY="http://host:port" 
client = Client()
while 1: 
    for i in range(1, 29):
        data = open('data/data%d.txt'%(i), 'r', encoding='utf-8').read()
        
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[{"role": "user", "content": "Tạo 100 câu hỏi ngẫu nhiên Từ đoạn văn sau:\n" + data}],
            max_tokens=40000
        )
        questions = response.choices[0].message.content.split('\n')
        true_questions = []
        for qs in questions:
            if(len(qs)>0):
                if qs[-1] == '?':
                    true_questions.append(qs) 
                    
        training_data = open('training_data.txt', 'a', encoding='utf-8')
        for qs in true_questions:
            response = client.chat.completions.create(
                model="gpt-3.5-turbo",
                messages=[{"role": "user", "content": "Trả lời câu hỏi sau:" + qs + " từ đoạn văn: " + data}],
                max_tokens=40000
            )
            answer = response.choices[0].message.content
            if (len(answer) > 0) and (answer[0] != '流'):
                training_data.write('<qs>' + qs + '\n' + '<as>' + answer + '\n')
        training_data.close()
```
